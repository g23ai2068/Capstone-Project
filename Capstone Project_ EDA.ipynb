# EDA for Calendar.csv

```python
#Using read_csv method to load the csv file into Pandas Dataframe
calendar = pd.read_csv('/content/drive/MyDrive/Capstone Project/calendar.csv')
calendar
```

```python
#Using shape method to check the number of rows and columns
print('Number of rows:', calendar.shape[0])
print('Number of rows:', calendar.shape[1])
```

```python
#Calculating the summary statistics to analyze and gain insights into the data values present in the calendar CSV file
calendar.describe()
```

```python
# Generating an informational table of the calendar CSV file in order to better understand the data types, null count of each column.
calendar.info()
```

```python
# Listing all the column names present in Calendar CSV
calendar.columns
```

```python
# Identifying columns that contain null values in the dataset.
calendar.isnull().sum()
```

```python
# All the values in 'adjusted_price' columns are null hence we can completely get rid of this column
# Dropping 'adjusted_price' from the csv
calendar1 = calendar.drop(columns=['adjusted_price'])
calendar1
```

```python
#Replacing t and f values in available column with True and False
calendar1['available'].replace(['t'], 'True', inplace=True)
calendar1['available'].replace(['f'], 'False', inplace=True)
calendar1
```
